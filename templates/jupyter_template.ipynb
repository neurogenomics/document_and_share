## Render

The following will convert this notebook into an html file.

### 1. In command line, activate your conda env (e.g. `pyre`). 

`conda activate pyre`  

**Note**: Depending on which version of conda you're using, you may need to enter `source activate <env_name>` instead.


### 2. Render the notebook: 

`jupyter nbconvert --to html_toc --execute templates/jupyter_template.ipynb --no-input`  


### 3. Push to GitHub

Push the new html file to GitHub.

`git add .`
`git commit -m "Some notes on updates"`
`git push`

### 4. Share the link 

Once GitHub Pages has updated (this can take a while if it's the first time you've added a given html file), you can infer the link by adding the path to the file (within the GitHub repo) to the url prefix (e.g. https://neurogenomics.github.io/document_and_share/): 

`https://neurogenomics.github.io/document_and_share/templates/jupyter_template.html` 

You can also embed the link directly into this Jupyter notebook:

[>> Results online here <<](https://neurogenomics.github.io/document_and_share/templates/jupyter_template.html)


<hr>

### Options

Here are the optional flags being used: 

- `--to html_toc`: Exports notebook at an html file with a table of contents (toc).
- `--execute`: This is only necessary when the notebook contains interactive plotly plots. 
- `--no-input`: This will hide all the code in the output file.

For a full list of export options, enter: 

`jupyter nbconvert -h`
